```python
# 哪些資料有高雄？
import os, sys
import xmltodict
dirs = os.listdir( './example' )
for file in dirs:
    if(file != 'sample.xml' and file !='__MACOSX'):
        with open('./example/'+file ,encoding = 'UTF-8') as fd:           
            doc = dict(xmltodict.parse(fd.read()))
            locationsName = doc['cwbopendata']['dataset']['locations']['locationsName']
            if('高雄' in locationsName ):
                print(file)
                print(locationsName)
        fd.close()
```
